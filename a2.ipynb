```python
"""
Task 1
- Create a function that searches for all occurrences of a value (find_val) in a given list (lst) and replaces them with another value (replace_val).
- lst must be a list.
- Return the modified list.
# Task 2
# Invoke the function "find_and_replace" using the following scenarios:
# - [1, 2, 3, 4, 2, 2], 2, 5
# - ["apple", "banana", "apple"], "apple", "orange"
"""

def find_and_replace(lst, find_val, replace_val):
    if not isinstance(lst,list):
        print("It is not a list. No operation is done")
    else:
        for i in range(len(lst)):
            if lst[i]==find_val:
                lst[i]=replace_val    
    return lst

lst1=[1, 2, 3, 4, 2, 2]
find_val1=2
replace_val1=5

lst2=["apple", "banana", "apple"]
find_val2="apple"
replace_val2="orange"

print(find_and_replace(lst1,find_val1,replace_val1))
print(find_and_replace(lst2,find_val2,replace_val2))

lst3="abcdefg"
find_val3="a"
replace_val3="b"

print(find_and_replace(lst3,find_val3,replace_val3))
```
