```python
"""
Task 1
- Create a function that would swap the value of x and y using only x and y as variables.
- x and y must be numeric.
- Return -1 if x and y is not numeric, and
- print the swapped values if both x and y are numeric.
# Task 2
# Invoke the function "swap" using the following scenarios:
# - "Apple", 10
# - 9, 17
"""
```

```python
#solution 1: 
#no return of the swapped value, since the question only asks for printing them.
#So this function can not be really used for swapping 2 variables.  

def swap(x,y):
    if (isinstance(x, (int, float,complex)) and isinstance(y, (int, float,complex))):
        x=x+y
        y=x-y
        x=x-y
        print (x,y)
        return
    return -1

print(swap("Apple", 10))
swap(9,17)
```

```python
#solution 2:
#really swap 2 variables when they are numeric.
#since this is a function to make the swapping done, return (y,x) directly is faster than changing x, y's value and return (x,y)

def swap(x,y):
    if (isinstance(x, (int, float,complex)) and isinstance(y, (int, float,complex))):
        return (y,x)
    return -1

a1="apple"
a2=10
a3=9
a4=17

if swap(a1,a2)==-1:
    print("At least 1 input variable is not numeric, not swapped")
else: 
    a1,a2=swap(a1,a2)[:]
    print(a1,a2)
    
if swap(a3,a4)==-1:
    print("At least 1 input variable is not numeric, not swapped")
else: 
    a3,a4=swap(a3,a4)[:]
    print(a3,a4)
```
